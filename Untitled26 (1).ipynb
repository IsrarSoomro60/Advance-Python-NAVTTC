```python
def greet(count):
    if count==0:
        return
    print("Hello")
    greet(count -1)
greet(10)
```

```python
def fib(n):
    if n<=0:
        return
    elif n ==1:
        return 1
    elif n ==2:
        return 2
    return fib(n-1) + fib(n-2)
fib(9)
```

## Check diagonal ##

```python
import numpy as np
```

```python
def right_diagonal(arr):
    for i in range(len(arr)-1):
        if arr[i][i] != arr[i+1][i+1]:
            return False
    return True
```

```python
right_diagonal(np.array([[10,2,3],
                         [4,10,5],
                         [6,1,1]]))
```

## Check Equal Row from 2dArray 

```python
import numpy as np

def check_row(arr, row_index):
    row = arr[row_index]   # jis row ko check karna hai
    first_value = row[0]
    
    for val in row:
        if val != first_value:
            return False
    print("Row", row_index, "all values are equal:", row)
    return True


# Example usage
check_row(np.array([[10, 2, 3],
                    [5, 5, 5],
                    [7, 8, 7]]), 1) # Row 1 -> True
# print(check_row)
```

## Create a function that takes 2d array as parameter input and return True if values are equal ##

```python
def column_values(arr):
    for i in range(len(arr)):
        for j in range(len(arr[i]-1)):
            if arr[j][i] != arr[j+1][i]:
                return False
    return True
```

```python
column_values(np.array([[10,2,3],
                         [10,10,5],
                         [10,1,1]]))
```

```python
class Chair:
    def __init__(self, num_legs, material, color):
        self.num_legs = num_legs
        self.material = material
        self.color = color
    def describe(self):
        print(f"A {self.color} {self.material} chair with {self.num_legs} legs")

chair1= Chair(4, "wood", "brown")
chair1= Chair(3, "Plastic", "White")
chair1.describe()
```
