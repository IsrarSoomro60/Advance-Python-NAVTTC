## Mutable and Immutable 

```python
data = "hello world "
print(data.strip(" "))
```

```python
lang = "I like java"
print(lang.replace("java", "Python"))
print(lang.replace(" ", ""))
```

```python
words= ["I", "love", "Python"]
print("".join(words))
```

```python
fvrt = "I love Python"
print(fvrt.split())
print(fvrt.split("v"))
```
