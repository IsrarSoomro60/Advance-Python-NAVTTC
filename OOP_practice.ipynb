# create a function that checks if all elements of 1d array are equal.

```python
import numpy as np
```

```python
def check_row(arr):
    temp = arr[0]
    for i in range(1,len(arr)):
        if temp != arr[i]:
            return False,0
    return True,arr[0]
    
```

```python
check_row(np.array([1,1,1,1,1]))
```

```python
def check_row2(arr):
    for i in range(len(arr)-1):
        if arr[i] != arr[i+1]:
            return False,0
    return True,arr[0]
```

```python
check_row2(np.array([1,0,1,1,1]))
```

# create a function that checks if all elements of any row of a 2d array is equal.

```python
def check_2d(arr):
    for i in range(len(arr)):
        temp = arr[i][0]
        # print(temp)
        flag = True
        for j in range(1,len(arr[i])):
            # print(f'remaining values: {arr[i][j]}')
            if temp != arr[i][j]:
                flag = False
        if flag:
            return True
    return False
```

```python
check_2d(np.array([[1,0,1],
                    [1,2,1],
                   [10,10,10]]))
```

```python
class Student:
    def __init__(self,name,grade,marks):
        self.name = name
        self.grade = grade
        self.marks = marks
    def average(self):
        return np.average(self.marks)
    def has_passed(self):
        print(f'average: {self.average()}')
        if self.average() > 50:
            return True
        else:
            return False
```

```python
student1 = Student('Ali','A',np.array([10,20,100]))
```

```python
print(student1.average())
average_marks = student1.average()
average_marks
```

```python
flag = student1.has_passed()
flag
```

```python
print(student1.has_passed(student1.average()))
```

```python
class BankAccount:
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount

    def withdraw(self,amount):
        self.balance -= amount

    def get_balance(self):
        print(f'your current balance is {self\
        .balance}')
```

```python
myaccount1 = BankAccount('owner',1000)
myaccount1.deposit(5000)
myaccount1.withdraw(2000)
myaccount1.get_balance()
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = {}
    def add_item(self,name, price):
        self.items[name] = price
    def remove_item(self,name):
        self.items.pop(name)
    def  get_total(self):
        return sum(self.items.values())
```

```python
cart1 = ShoppingCart()
cart1.add_item('Surf Excel',1000)
cart1.add_item('Olpers',500)
cart1.add_item('tissue box',100)
cart1.remove_item('tissue box')
```

```python
cart1.items
cart1.get_total()
```

```python
class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2* self.length + 2 * self.width
```

```python
rect1 = Rectangle(100,200)
print(f'perimeter = {rect1.perimeter()}')
print(f'area = {rect1.area()}')
```
