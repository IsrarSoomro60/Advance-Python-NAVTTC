```python
def greet(count):
    if count==0:
        return
    print("Hello")
    greet(count -1)
greet(10)
```

```python
def fib(n):
    if n<=0:
        return
    elif n ==1:
        return 1
    elif n ==2:
        return 2
    return fib(n-1) + fib(n-2)
fib(9)
```

## Check diagonal ##

```python
import numpy as np
```

```python
def right_diagonal(arr):
    for i in range(len(arr)-1):
        if arr[i][i] != arr[i+1][i+1]:
            return False
    return True
```

```python
right_diagonal(np.array([[10,2,3],
                         [4,10,5],
                         [6,1,1]]))
```

## Check Equal Row from 2dArray 

```python
import numpy as np

def check_row(arr, row_index):
    row = arr[row_index]   # jis row ko check karna hai
    first_value = row[0]
    
    for val in row:
        if val != first_value:
            return False
    print("Row", row_index, "all values are equal:", row)
    return True


# Example usage
check_row(np.array([[10, 2, 3],
                    [5, 5, 5],
                    [7, 8, 7]]), 1) # Row 1 -> True
# print(check_row)
```

## Create a function that takes 2d array as parameter input and return True if values are equal ##

```python
def column_values(arr):
    n_rows, n_cols = arr.shape
    for col in range(n_cols):
        first_value = arr[0, col]
        for row in range(1, n_rows):
            if arr[row, col] != first_value:
                return False
    return True

column_values(np.array([[10, 2, 3],
                         [10, 5, 5],
                         [10, 8, 7]]))
```
