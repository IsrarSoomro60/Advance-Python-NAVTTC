```python
class Animal:
    def eat(self):
        print("This animal eats food")
    def sleep(self):
        print("This animal sleeps")

class Dog(Animal):
    def bark(self):
        print("The dog barks")

class Cat(Animal):
    def meow(self):
        print("The cat meows")

my_dog =Dog()
my_dog.bark()
my_dog.eat()
my_dog.sleep()
# my_cat = Cat()
my_cat.meow()
```

```python
# Parent Class
class User:
    def __init__(self, name, email):
        self.name = name
        self.email = email
    
    def login(self):
        print(f"{self.name} logged in.")
    
    def logout(self):
        print(f"{self.name} logged out.")

# Child Class 1: Customer
class Customer(User):
    def __init__(self, name, email, address):
        super().__init__(name, email)
        self.address = address
    
    def place_order(self, product):
        print(f"{self.name} placed an order for {product}.")

# Child Class 2: Seller
class Seller(User):
    def __init__(self, name, email, shop_name):
        super().__init__(name, email)
        self.shop_name = shop_name
    
    def add_product(self, product):
        print(f"{self.name} added {product} to shop {self.shop_name}.")

# Child Class 3: Admin
class Admin(User):
    def __init__(self, name, email):
        super().__init__(name, email)
    
    def remove_user(self, user):
        print(f"Admin {self.name} removed user {user.name} from system.")

# Testing
customer1 = Customer("Israr", "israr@email.com", "Karachi")
seller1 = Seller("Ahmed", "ahmed@shop.com", "TechStore")
admin1 = Admin("Sana", "sana@admin.com")

customer1.login()
customer1.place_order("Laptop")
seller1.add_product("Smartphone")
admin1.remove_user(customer1)
```
